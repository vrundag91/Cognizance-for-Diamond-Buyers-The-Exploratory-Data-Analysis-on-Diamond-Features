```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib import style
plt.style.use(['dark_background'])

import seaborn as sns
sns.set(color_codes=True)
sns.set(font_scale=1.2)
```

```python
d = sns.load_dataset('diamonds')
```

```python
d.shape
```

```python
d.head(30)
```

```python
df = d
```

```python
df.shape
```

```python
d=df.sample(1000)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(25, 15), sharey=True)
fig.suptitle('Correlation')


sns.scatterplot(ax=axes[0,0], data=d, x=d.carat, y=d.price, hue = "cut", palette='gray');
axes[0,0].set_title("(a) caret v/s price");

sns.scatterplot(ax=axes[0,1], x=d.depth, y=d.price, data=d, hue = "cut", palette='gray');
axes[0,1].set_title("(b) total Depth v/s price");

sns.scatterplot(ax=axes[0,2], x=d.table, y=d.price, data=d, hue = "cut", palette='gray');
axes[0,2].set_title("(c) table v/s price");

sns.scatterplot(ax=axes[1,0], data=d, x=d.x, y=d.price, hue = "cut", palette='gray');
axes[1,0].set_title("(d) x(Length) v/s price");

sns.scatterplot(ax=axes[1,1], x=d.y, y=d.price, data=d, hue = "cut", palette='gray');
axes[1,1].set_title("(e) y(Width) v/s price");

sns.scatterplot(ax=axes[1,2], x=d.z, y=d.price, data=d, hue = "cut", palette='gray');
axes[1,2].set_title("(f) z(Depth) v/s price");
```

```python

fig, axes = plt.subplots(2, 3, figsize=(25, 15), sharey=True)
fig.suptitle('Outliers')

sns.boxplot(ax=axes[0,0], data=df, x='carat', palette='gray');
axes[0,0].set_title("(a) caret Outliers");

sns.boxplot(ax=axes[0,1], data=df, x='depth', palette='gray');
axes[0,1].set_title("(b) depth Outliers");

sns.boxplot(ax=axes[0,2], data=df, x='table', palette='gray');
axes[0,2].set_title("(c) table outliers");

sns.boxplot(ax=axes[1,0], data=df, x='x', palette='gray');
axes[1,0].set_title("(d) x(length) outliers");

sns.boxplot(ax=axes[1,1], data=df, x='y', palette='gray');
axes[1,1].set_title("(e) y(Width) outliers");

sns.boxplot(ax=axes[1,2], data=df, x='z', palette='gray');
axes[1,2].set_title("(f) z(Depth) outliers");
```
